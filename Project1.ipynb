```python
# ============================================
# MERGE CORRECTION with cruz_list
# ============================================

import pandas as pd

# 1. Reload original data
df = pd.read_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv")
df_cruz = pd.read_csv("/content/drive/MyDrive/Project1_VI/cruz_list.csv", sep=';')

print("=== DATA LOADED ===")
print(f"Completed scholarships: {len(df)}")
print(f"Cruz list scholarships: {len(df_cruz)}")

# 2. Clean IDs to ensure matching
# Convert to string and remove spaces
df['grant_id_clean'] = df['grant_id'].astype(str).str.strip()
df_cruz['grant_number_clean'] = df_cruz['grant_number'].astype(str).str.strip()

print("\n=== CLEANED IDs ===")
print("Examples grant_id_clean:", df['grant_id_clean'].head(3).tolist())
print("Examples grant_number_clean:", df_cruz['grant_number_clean'].head(3).tolist())

# 3. Create a list of IDs in Cruz
cruz_ids = set(df_cruz['grant_number_clean'].unique())
print(f"\n{len(cruz_ids)} unique IDs in Cruz list")

# 4. Mark scholarships that are in the Cruz list
df['in_cruz_list_fixed'] = df['grant_id_clean'].isin(cruz_ids)

print("\n=== CORRECTED MERGE RESULTS ===")
print(f"✅ In Cruz list: {df['in_cruz_list_fixed'].sum()} ({df['in_cruz_list_fixed'].mean()*100:.1f}%)")
print(f"❌ NOT in Cruz list: {(~df['in_cruz_list_fixed']).sum()} ({(~df['in_cruz_list_fixed']).mean()*100:.1f}%)")

# 5. Compare with old column
print("\n=== OLD vs NEW COMPARISON ===")
comparison = pd.DataFrame({
    'Old (in_cruz_list)': [
        (df['in_cruz_list'] == True).sum(),
        (df['in_cruz_list'] == False).sum(),
        df['in_cruz_list'].isna().sum()
    ],
    'New (in_cruz_list_fixed)': [
        df['in_cruz_list_fixed'].sum(),
        (~df['in_cruz_list_fixed']).sum(),
        0
    ]
}, index=['True', 'False', 'NaN'])
print(comparison)

# 6. Replace old column
df['in_cruz_list'] = df['in_cruz_list_fixed']

# 7. Clean temporary columns
df = df.drop(columns=['grant_id_clean', 'in_cruz_list_fixed'])

# 8. Save
df.to_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv", index=False)
print("\n✅ Corrected dataset saved!")
```

```python
# ============================================
# Creation of has_flagged_words (Q4)
# ============================================

df_flagged = pd.read_csv("/content/drive/MyDrive/Project1_VI/flagged_words_trump_admin.csv")

print("=== FLAGGED KEYWORDS ===")
display(df_flagged.head())

# Extract the keywords (adjust the column name if needed)
if 'word' in df_flagged.columns:
    flagged_words = df_flagged['word'].str.lower().tolist()
elif 'keyword' in df_flagged.columns:
    flagged_words = df_flagged['keyword'].str.lower().tolist()
else:
    # If you don't know the column name
    print("Available columns:", df_flagged.columns.tolist())
    flagged_words = df_flagged.iloc[:, 0].str.lower().tolist()  # Take the first column

print(f"\n{len(flagged_words)} keywords: {flagged_words}")

# Detection function
def contains_flagged_words(text, keywords):
    if pd.isna(text):
        return False
    text_lower = str(text).lower()
    return any(keyword in text_lower for keyword in keywords)

# Create combined text
df['combined_text'] = (
    df['project_title'].fillna('') + ' ' +
    df['abstract'].fillna('')
)

# Detection
df['has_flagged_words'] = df['combined_text'].apply(
    lambda x: contains_flagged_words(x, flagged_words)
)

# Statistics
print(f"\n=== Q4 RESULTS ===")
print(f"✅ With keywords: {df['has_flagged_words'].sum()} ({df['has_flagged_words'].mean()*100:.1f}%)")
print(f"❌ Without keywords: {(~df['has_flagged_words']).sum()} ({(~df['has_flagged_words']).mean()*100:.1f}%)")

# Examples
print("\n=== EXAMPLES OF SCHOLARSHIPS WITH KEYWORDS ===")
display(df[df['has_flagged_words']][['grant_id', 'org_name', 'project_title']].head(3))

# Clean and save
df = df.drop(columns=['combined_text'])
df.to_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv", index=False)
print("\n✅ has_flagged_words column added and saved!")
```

```python
# ============================================
# Q1 - OPTION C: Combined Map + Bar Chart
# ============================================

# 1. Map (same as Option B)
map_chart = alt.Chart(states).mark_geoshape(
    stroke='white',
    strokeWidth=0.5
).encode(
    color=alt.Color('num_grants:Q',
                    scale=alt.Scale(scheme='orangered', type='sqrt'),
                    title='Grants',
                    legend=None),
    tooltip=[
        alt.Tooltip('org_state:N', title='State'),
        alt.Tooltip('num_grants:Q', title='Canceled Grants', format=',d')
    ]
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(state_counts, 'id', ['num_grants', 'org_state'])
).project(
    type='albersUsa'
).properties(
    width=600,
    height=350,
    title='Geographic Distribution of Cancellations'
)

# 2. Bar Chart Top 10
top10_states = state_counts.nlargest(10, 'num_grants')

bar_chart = alt.Chart(top10_states).mark_bar().encode(
    x=alt.X('num_grants:Q',
            title='Number of Canceled Grants',
            axis=alt.Axis(format='d')),
    y=alt.Y('org_state:N',
            title='State',
            sort='-x'),
    color=alt.Color('num_grants:Q',
                    scale=alt.Scale(scheme='orangered'),
                    legend=None),
    tooltip=[
        alt.Tooltip('org_state:N', title='State'),
        alt.Tooltip('num_grants:Q', title='Grants', format=',d')
    ]
).properties(
    width=600,
    height=250,
    title='Top 10 Most Affected States'
)

# 3. Combine vertically
chart_q1_combined = alt.vconcat(
    map_chart,
    bar_chart
).resolve_scale(
    color='independent'
)

chart_q1_combined
```

```python
# ============================================
# Q2 - CLEAN AND CORRECTED LOLLIPOP CHART
# ============================================

import pandas as pd
import altair as alt

# Load data
df = pd.read_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv")

# Aggregation by institution
institution_stats = df.groupby('org_name').agg({
    'grant_id': 'count',
    'budget_preferred': 'sum'
}).rename(columns={
    'grant_id': 'num_grants',
    'budget_preferred': 'total_budget'
}).reset_index().sort_values('num_grants', ascending=False)

# Statistics
mean_grants = institution_stats['num_grants'].mean()
median_grants = institution_stats['num_grants'].median()

print(f"=== Q2 STATS ===")
print(f"Mean: {mean_grants:.2f} grants/institution")
print(f"Top 1: {institution_stats.iloc[0]['org_name']} ({institution_stats.iloc[0]['num_grants']} grants)")

# Top 15 for readability
top_n = 15
top_institutions = institution_stats.head(top_n).copy()

# Shorten overly long names
def shorten_name(name, max_length=55):
    """Shortens overly long names"""
    if len(name) > max_length:
        return name[:max_length-3] + '...'
    return name

top_institutions['org_name_display'] = top_institutions['org_name'].apply(shorten_name)

# ============================================
# LOLLIPOP CHART CONSTRUCTION
# ============================================

# 1. LINES (stems) - from 0 up to the point
lines = alt.Chart(top_institutions).mark_rule(
    color='#d62728',
    strokeWidth=2.5,
    opacity=0.8
).encode(
    x=alt.X('start:Q'),
    x2=alt.X2('num_grants:Q'),
    y=alt.Y('org_name_display:N',
            sort='-x',
            title=None,
            axis=alt.Axis(
                labelLimit=450,
                labelFontSize=11,
                labelPadding=10
            ))
).transform_calculate(
    start='0'  # Lines start at 0
)

# 2. POINTS (lollipops)
points = alt.Chart(top_institutions).mark_circle(
    size=200,
    color='#d62728',
    opacity=1
).encode(
    x=alt.X('num_grants:Q',
            title='Number of Cancelled Grants',
            scale=alt.Scale(domain=[0, top_institutions['num_grants'].max() * 1.12]),
            axis=alt.Axis(
                format='d',
                grid=True,
                gridOpacity=0.2,
                tickCount=8
            )),
    y=alt.Y('org_name_display:N',
            sort='-x'),
    tooltip=[
        alt.Tooltip('org_name:N', title='Institution'),
        alt.Tooltip('num_grants:Q', title='Cancelled Grants', format=',d'),
        alt.Tooltip('total_budget:Q', title='Lost Budget', format='$,.0f')
    ]
)

# 3. LABELS with values
text_values = alt.Chart(top_institutions).mark_text(
    align='left',
    baseline='middle',
    dx=10,
    fontSize=11,
    fontWeight='bold',
    color='#333333'
).encode(
    x=alt.X('num_grants:Q'),
    y=alt.Y('org_name_display:N', sort='-x'),
    text=alt.Text('num_grants:Q', format='d')
)

# 4. REFERENCE LINE: Mean
mean_data = pd.DataFrame({'mean_value': [mean_grants]})

mean_line = alt.Chart(mean_data).mark_rule(
    strokeDash=[8, 4],
    color='#666666',
    strokeWidth=2,
    opacity=0.7
).encode(
    x='mean_value:Q'
)

# 5. ANNOTATION for the mean (well-positioned)
mean_text = alt.Chart(pd.DataFrame({
    'x': [mean_grants],
    'y': [top_institutions.iloc[0]['org_name_display']],  # At the top of the chart
    'label': [f'Mean: {mean_grants:.1f}']
})).mark_text(
    align='center',
    baseline='bottom',
    dy=-8,
    fontSize=11,
    fontWeight='bold',
    color='#666666',
    fontStyle='italic'
).encode(
    x=alt.X('x:Q'),
    y=alt.Y('y:N'),
    text='label:N'
)

# ============================================
# COMBINE ALL LAYERS
# ============================================

chart_q2_lollipop = alt.layer(
    lines,           # 1. Lines first (background)
    mean_line,       # 2. Mean line
    points,          # 3. Points on top
    text_values,     # 4. Numerical values
    mean_text        # 5. Mean label
).properties(
    width=750,
    height=550,
    title={
        'text': f'Top {top_n} Institutions Affected by Number of Cancelled Grants',
        'subtitle': f'Mean: {mean_grants:.1f} grants per institution | Total: {institution_stats["num_grants"].sum()} grants across {len(institution_stats)} institutions',
        'fontSize': 15,
        'subtitleFontSize': 11,
        'anchor': 'start',
        'dx': 10
    }
).configure_view(
    strokeWidth=0
).configure_axis(
    labelFontSize=11,
    titleFontSize=12,
    domainWidth=1,
    domainColor='#cccccc',
    tickColor='#cccccc'
)

chart_q2_lollipop
```

```python
# ============================================
# Q4 - FIX: WORD BOUNDARY SEARCH
# ============================================

import pandas as pd
import re

# Reload data
df = pd.read_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv")

# Load flagged words
df_flagged = pd.read_csv("/content/drive/MyDrive/Project1_VI/flagged_words_trump_admin.csv")

# Get flagged words
if 'word' in df_flagged.columns:
    flagged_words = df_flagged['word'].str.lower().tolist()
elif 'keyword' in df_flagged.columns:
    flagged_words = df_flagged['keyword'].str.lower().tolist()
else:
    flagged_words = df_flagged.iloc[:, 0].str.lower().tolist()

print(f"=== FLAGGED WORDS ===")
print(f"Total: {len(flagged_words)}")
print(f"Words: {flagged_words}")

# ============================================
# IMPROVED: Word boundary matching
# ============================================

def contains_flagged_words_strict(text, keywords):
    """
    Check if text contains any flagged keywords using WORD BOUNDARIES.
    This avoids false positives like 'trans' matching 'transportation'.
    """
    if pd.isna(text):
        return False

    text_lower = str(text).lower()

    for keyword in keywords:
        # Use word boundaries \b to match whole words only
        # \b matches position between word and non-word character
        pattern = r'\b' + re.escape(keyword) + r'\b'

        if re.search(pattern, text_lower):
            return True

    return False

# Create combined text
df['combined_text'] = (
    df['project_title'].fillna('') + ' ' +
    df['abstract'].fillna('')
)

# Apply STRICT detection
print("\n=== APPLYING STRICT WORD BOUNDARY DETECTION ===")
df['has_flagged_words_strict'] = df['combined_text'].apply(
    lambda x: contains_flagged_words_strict(x, flagged_words)
)

# Compare OLD vs NEW
print("\n=== COMPARISON: OLD vs NEW ===")
if 'has_flagged_words' in df.columns:
    print(f"OLD (substring search):")
    print(f"  WITH flagged words: {df['has_flagged_words'].sum()} ({df['has_flagged_words'].mean()*100:.1f}%)")

print(f"\nNEW (word boundary search):")
print(f"  WITH flagged words: {df['has_flagged_words_strict'].sum()} ({df['has_flagged_words_strict'].mean()*100:.1f}%)")
print(f"  WITHOUT flagged words: {(~df['has_flagged_words_strict']).sum()} ({(~df['has_flagged_words_strict']).mean()*100:.1f}%)")

# Show examples of changes
print("\n=== EXAMPLES OF CORRECTIONS ===")
changed = df[df['has_flagged_words'] != df['has_flagged_words_strict']]
print(f"Number of grants reclassified: {len(changed)}")

if len(changed) > 0:
    print("\nExamples of FALSE POSITIVES (old method incorrectly flagged these):")
    false_positives = changed[changed['has_flagged_words'] & ~changed['has_flagged_words_strict']]
    for idx, row in false_positives.head(5).iterrows():
        print(f"\n  Grant: {row['grant_id']}")
        print(f"  Title: {row['project_title'][:80]}...")
        # Show which word was falsely detected
        for word in flagged_words:
            if word in row['combined_text'].lower():
                print(f"  → False match on '{word}'")
                break

# Count word occurrences with STRICT matching
print("\n=== FLAGGED WORDS FREQUENCY (STRICT) ===")
word_counts_strict = {}
for word in flagged_words:
    pattern = r'\b' + re.escape(word) + r'\b'
    count_title = df['project_title'].str.lower().str.count(pattern).sum()
    count_abstract = df['abstract'].fillna('').str.lower().str.count(pattern).sum()
    total_count = count_title + count_abstract
    if total_count > 0:
        word_counts_strict[word] = total_count

word_counts_sorted = sorted(word_counts_strict.items(), key=lambda x: x[1], reverse=True)
print("Most frequent flagged words (strict matching):")
for word, count in word_counts_sorted[:15]:
    print(f"  '{word}': {count} occurrences")

# Replace old column with new one
df = df.drop(columns=['has_flagged_words', 'combined_text'], errors='ignore')
df = df.rename(columns={'has_flagged_words_strict': 'has_flagged_words'})

# Save
df.to_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv", index=False)
print("\n✅ Dataset updated with strict word boundary matching!")
```

```python
# ============================================
# Q4 - VERTICAL STACKED VERSION
# ============================================

import pandas as pd
import altair as alt

# Load data
df = pd.read_csv("/content/drive/MyDrive/Project1_VI/merged_clean_ready.csv")

# Statistics
total_grants = len(df)
with_flagged = df['has_flagged_words'].sum()
without_flagged = total_grants - with_flagged

budget_with = df[df['has_flagged_words']]['budget_preferred'].sum()
budget_without = df[~df['has_flagged_words']]['budget_preferred'].sum()

# Prepare data
summary_data = pd.DataFrame({
    'Category': ['With Flagged Words', 'Without Flagged Words'],
    'Grants': [with_flagged, without_flagged],
    'Grants_pct': [with_flagged/total_grants*100, without_flagged/total_grants*100],
    'Budget_M': [budget_with/1_000_000, budget_without/1_000_000],
    'Budget_pct': [budget_with/(budget_with+budget_without)*100,
                   budget_without/(budget_with+budget_without)*100]
})

# Chart 1: Grants
chart1 = alt.Chart(summary_data).mark_bar(
    cornerRadiusEnd=4,
    height=50
).encode(
    x=alt.X('Grants:Q',
            title='Number of Grants',
            axis=alt.Axis(format='d')),
    y=alt.Y('Category:N',
            title=None,
            sort='-x',
            axis=alt.Axis(labelLimit=250)),
    color=alt.Color('Category:N',
                    scale=alt.Scale(
                        domain=['With Flagged Words', 'Without Flagged Words'],
                        range=['#ff7f0e', '#1f77b4']
                    ),
                    legend=None),
    tooltip=[
        alt.Tooltip('Category:N', title='Category'),
        alt.Tooltip('Grants:Q', title='Grants', format=',d'),
        alt.Tooltip('Grants_pct:Q', title='%', format='.1f')
    ]
).properties(
    width=650,
    height=150,
    title='Number of Grants'
)

text1 = chart1.mark_text(align='left', dx=5, fontSize=12, fontWeight='bold').encode(
    text=alt.Text('Grants:Q', format='d'),
    color=alt.value('black')
)

# Chart 2: Budget
chart2 = alt.Chart(summary_data).mark_bar(
    cornerRadiusEnd=4,
    height=50
).encode(
    x=alt.X('Budget_M:Q',
            title='Budget Lost (millions $)',
            axis=alt.Axis(format='$.0f')),
    y=alt.Y('Category:N',
            title=None,
            sort='-x',
            axis=alt.Axis(labelLimit=250)),
    color=alt.Color('Category:N',
                    scale=alt.Scale(
                        domain=['With Flagged Words', 'Without Flagged Words'],
                        range=['#2ca02c', '#1f77b4']
                    ),
                    legend=None),
    tooltip=[
        alt.Tooltip('Category:N', title='Category'),
        alt.Tooltip('Budget_M:Q', title='Budget (M$)', format='$,.0f'),
        alt.Tooltip('Budget_pct:Q', title='%', format='.1f')
    ]
).properties(
    width=650,
    height=150,
    title='Budget Lost'
)

text2 = chart2.mark_text(align='left', dx=5, fontSize=12, fontWeight='bold').encode(
    text=alt.Text('Budget_M:Q', format='$.0f'),
    color=alt.value('black')
)

# Stack vertically
chart_q4_vertical = alt.vconcat(
    (chart1 + text1),
    (chart2 + text2),
    spacing=30
).properties(
    title={
        'text': 'Q4: Correlation between Cancelled Grants and Flagged Words',
        'subtitle': f'{with_flagged} grants (84.6%) contain flagged words | ${budget_with/1_000_000:.0f}M (90.1%) of budget',
        'fontSize': 16,
        'subtitleFontSize': 11
    }
).configure_view(
    strokeWidth=0
).resolve_scale(
    color='independent'
)

chart_q4_vertical
```
